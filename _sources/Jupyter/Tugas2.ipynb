```python
import numpy as np
A = np.array([[1,6,2],[2,4,3],[3,1,4]])
B = np.array([[2,5,4],[4,1,3],[4,8,3]])
C = np.array([[3,6],[9,4],[9,3]])
D = np.array([[8,4,2],[3,1,6],[4,5,7]])
E = A+B
F = 2*A
G = A@C
H = B@C
I = A@H
J = A@B
K = J@C


print(A)
print('\n')
print(B)
print('\n')
print(C)
print('\n')
print(D)
print('\n')
print(E)
print('\n')
print(F)
print('\n')
print(G)
print('\n')
print(H)
print('\n')
print(I)
print('\n')
print(J)
print('\n')
print(K)
print('\n')
```
